# **Relation**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11) # con np linespace se crea un array de 0 a 5 con 11 numeros
y=x**2
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')

plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y,'y') #el ultimo argumento es el color de la grafica
```

```python
plt.plot(x,y,'rx') #el ultimo argumento es el color de la grafica y la forma de los datos en este caso x pero puede ser o
```

```python
plt.plot(x,y,'rs-')
```

```python
plt.plot(x,y,'yD:')
```

```python
plt.hist(x) #hist son graficas de tipo  histogramas
```

# **Subplot**

```python
plt.subplot(1,2,1) # en orden 1 es el numero de filas, 2 el numero de columnas y 1 la columna que vamos a graficar para este caso un plot de linea
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1) # en el primer subplot hay dos gáficas.
plt.plot(x,y)
plt.plot(y,x)
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)# como se puede apreciar en esta ocasión cambiamos a dos filas y una sola columna
plt.plot(x,y)
plt.plot(y,x)
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

# **Método orientado a objetos**

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.1,0.5,0.5])#primeros dos valores ubicacion en e x ,y. Los segundos valores el tamaño de la grafica.
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])#primeros dos valores ubicacion en e x ,y. Los segundos valores el tamaño de la grafica.
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3]) # en este caso toma mucho sentido hacer la gráfica pequeña y cambiar su ubicación para que no oculte los otros datos
axes.plot(x,y)
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')# se cambia el fondo de color
```

# **Leyendas, etiquetas, títulos, tamaño**

```python
fig, axes = plt.subplots(1,2) # subplots son 1 fila y dos columnas en el lienzo
```

```python
y=np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(5,5)) # subplots son 1 fila y dos columnas en el lienzo, figsize nos da el tamaño de la grafica
axes[0].plot(x,y,label='$sin(x)$')#defino el label en el plot, con el simbolo $ se deja la notacion matematica
axes[0].set_title('Relacion X-Y')# Titulo del subplot
axes[0].set_xlabel('X')#Titulo de los ejes x y y
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y-X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y, label="sin(x)") # sin tener en cuenta el tipo objetos sino plot normal
plt.title('Relacion x-y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, label="sin(x)")
plt.title('Relacion x-y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(y,x, label="sin(x)")
plt.title('Relacion y-x')
plt.xlabel('Y')
plt.ylabel('X')
plt.legend()
```

## **Bar plot**

```python
plt.style.use('default')# seleccion del tipo de formato
```

```python
print(plt.style.available) # todos los formatos predefinidos por la libreria
```

```python
country=['INDIA','JAPON','MEXICO','COLOMBIA','GERMAY']
population=[1000,800,900,1000,300]
```

```python
plt.bar(country,population, width=0.5, color=['red','blue']) #width me da el ancho de las barras
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)#cambia el nombre de los paises. Con rotation puedo girar algo el nombre de los paises
plt.show()
```

```python
plt.barh(country,population,color=['#F39C12','blue','green','yellow','red']) # con barh se hacen horizontales las barras
plt.title('Country Population',fontsize=20, ha='center',color="black")
plt.xlabel('Numero de habitantes en miles',fontsize=15, ha='center')
plt.ylabel('Paises',fontsize=15, ha='center')
plt.yticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'))
plt.show()
```

# **Categóricos**

```python
sns.countplot(data=tips,x='day',hue='sex') # en los datos categoricos, lo que buscamos es contar por categorias.
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)# esta gráfica es exactamente igual a la anterior pero con dodge evitamos que se mezclen las categorias a la hora de visualizar
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a “beeswarm”

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,palette="husl")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True) #catplot, es el equivalente al displot, por defecto nos da uno pero lo podemos ir ajustando
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind="box") # aca en este con el kind se pueden usar todos los tipode de graficas mencionadas anteriormente
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,kind="box",col='time') # solamente con incluir col time nos muesta las variaciones adicionales
```
